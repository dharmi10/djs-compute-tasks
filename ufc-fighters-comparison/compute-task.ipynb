```python
#categorical values with mode
data['nickname'] = data['nickname'].fillna(data['nickname'].mode()[0])
data['stance'] = data['stance'].fillna(data['stance'].mode()[0])
data['date_of_birth'] = data['date_of_birth'].fillna(data['date_of_birth'].mode()[0])

#numeic values with mean/median
data['height_cm'] = data['height_cm'].fillna(data['height_cm'].mean())
data['reach_in_cm'] = data['reach_in_cm'].fillna(data['reach_in_cm'].mean())
data['weight_in_kg'] = data['weight_in_kg'].fillna(data['weight_in_kg'].median())

data.isnull().sum()
```

```python
print(f"Numerical columns: {len(data.select_dtypes(include=['number']).columns)}")
print(f"Text/Categorical columns: {len(data.select_dtypes(include=['object']).columns)}")
```

```python
data['date_of_birth'] = pd.to_datetime(data['date_of_birth'])
current_date = pd.to_datetime('today')
data['age'] = (datetime.now() - data['date_of_birth']).dt.days // 365
data.head()
```

```python
core_columns = ["age", "height_cm", "weight_in_kg", "reach_in_cm", 
                "wins", "losses", "draws"]

data[core_columns].hist(bins=20, figsize=(12,8), edgecolor="black")
plt.show()
```

```python
data['total_fights'] = data['wins'] + data['losses'] + data['draws']
data['win_rate'] = data['wins'] / data['total_fights']
data['loss_rate'] = data['losses'] / data['total_fights']
data['draw_rate'] = data['draws'] / data['total_fights']
data['win_loss_ratio'] = data['wins'] / (data['losses'] + 1)
data.head(15)
```
